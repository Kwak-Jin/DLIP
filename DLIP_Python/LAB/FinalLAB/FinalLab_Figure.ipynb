```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

data = np.array([
    [50, 0, 0, 0, 0],
    [0, 45, 1, 12, 0],
    [0, 2, 49, 4, 0],
    [0, 35, 2, 43, 0],
    [0, 0, 0, 0, 44]
])

labels = ['A', 'B', 'C', 'D', 'E']

# Create a DataFrame
DATA_CONFUSION = pd.DataFrame(data, index=labels, columns=labels)

# Plotting the heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(DATA_CONFUSION, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score, precision_score, recall_score, fbeta_score

# Data extracted from the document
confusion_matrix = np.array([
    [50, 0, 0, 0, 0],
    [0, 45, 1, 12, 0],
    [0, 2, 49, 4, 0],
    [0, 35, 2, 43, 0],
    [0, 0, 0, 0, 44]
])

# True labels and predicted labels (flattened from the confusion matrix)
true_labels = []
pred_labels = []
labels = ['A', 'B', 'C', 'D', 'E']

for i, label in enumerate(labels):
    for j, _ in enumerate(labels):
        true_labels.extend([label] * confusion_matrix[i, j])
        pred_labels.extend([labels[j]] * confusion_matrix[i, j])
beta_f2  = 2
beta_f0_5= 0.5
# Calculate precision, recall, F1, F2, and F0.5 scores
precision = precision_score(true_labels, pred_labels, labels=labels, average=None)
recall = recall_score(true_labels, pred_labels, labels=labels, average=None)
f1 = f1_score(true_labels, pred_labels, labels=labels, average=None)
f2 = fbeta_score(true_labels, pred_labels, beta = beta_f2,labels=labels, average=None)
f0_5 = fbeta_score(true_labels, pred_labels, beta = beta_f0_5,labels=labels, average=None)

# Plot the metrics
metrics = [precision, recall, f1, f2, f0_5]
metrics_names = ['Precision', 'Recall', 'F1 Score', 'F2 Score', 'F0.5 Score']
x = np.arange(len(labels))

fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(25, 5), sharey=True)
for ax, metric, name in zip(axes, metrics, metrics_names):
    ax.bar(x, metric)
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.set_title(name)

fig.suptitle('Classification Metrics for Each Class')
plt.show()
```
